```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tp=[]
tn=[]
fp=[]
fn=[]
with open("C:/Users/vatan/Downloads/no_update.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            tp.append(int(splitted_line[1]))
            tn.append(int(splitted_line[2]))
            fp.append(int(splitted_line[3]))
            fn.append(int(splitted_line[4]))
```

```python
print(len(tp))
print(len(tn))
print(len(fp))
print(len(fn))
```

```python
total_tp=sum(tp)
total_tn=sum(tn)
total_fp=sum(fp)
total_fn=sum(fn)
```

```python
cm_data = [[total_tp,total_fn], [total_fp, total_tn]]
y_labels=["Positive","Negative"]
x_labels=["Positive","Negative"]
labels=["TP","FP","FN","TN"]
plt.figure(figsize = (10,7))
plt.title('First Test Case Confusion Matrix')
sns.heatmap(cm_data, annot=True, cmap='Blues',fmt='d',yticklabels=y_labels,xticklabels=x_labels)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```
