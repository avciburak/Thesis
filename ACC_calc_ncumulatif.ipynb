```python
from matplotlib import pyplot as plt
plt.style.use("seaborn")
```

```python
acc_nu=[]
pre_nu=[]
rec_nu=[]
acc=0.0
pre=0.0
rec=0.0
tp=0
tn=0
fp=0
fn=0
with open("C:/Users/vatan/Downloads/no_update.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            tp=float(splitted_line[1])
            tn=float(splitted_line[2])
            fp=float(splitted_line[3])
            fn=float(splitted_line[4])
            
            try:
                acc=(float(tp)+float(tn))/(float(tp)+float(tn)+float(fn)+float(fp))
                acc_nu.append(acc)
            except ZeroDivisionError:
                pass
            
            try:
                pre=(float(tp))/(float(tp)+float(fp))
                pre_nu.append(pre)
            except ZeroDivisionError:
                pass
            try:
                rec=(float(tp))/(float(tp)+float(fn))
                rec_nu.append(rec)
            except ZeroDivisionError:
                pass
```

```python
x=list(range(len(acc_nu)))
plt.plot(x,acc_nu,color='r',label='no update')
#plt.plot(x,acc_ueff[0:15750],color='g',label='updated in each five frames')
plt.xlabel("Frame Number")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
