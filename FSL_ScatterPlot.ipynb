```python
from matplotlib import pyplot as plt
plt.style.use("seaborn")
```

```python
iou_nu=[]
scores_nu=[]
with open("C:/Users/vatan/Downloads/iou&scores.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            iou_str=splitted_line[3]
            scores_str=splitted_line[2]
            #if float(iou_str)>0.3:
            iou_nu.append(float(iou_str))
            scores_nu.append(float(scores_str))
```

```python
print(len(scores_nu))
print(len(iou_nu))
```

```python
zerozeropone=[]
zeroponetwo=[]
zeroptwothree=[]
zeropthreefour=[]
zeropfourfive=[]
zeropfivesix=[]
zeropsixseven=[]
zeropseveneight=[]
zeropeightnine=[]
zeropnineone=[]
for score in scores_nu:
    if score>=0 and score<0.1:
        zerozeropone.append(score)
    elif score>=0.1 and score<0.2:
        zeroponetwo.append(score)
    elif score>=0.2 and score<0.3:
        zeroptwothree.append(score)
    elif score>=0.3 and score<0.4:
        zeropthreefour.append(score)
    elif score>=0.4 and score<0.5:
        zeropfourfive.append(score)
    elif score>=0.5 and score<0.6:
        zeropfivesix.append(score)
    elif score>=0.6 and score<0.7:
        zeropsixseven.append(score)
    elif score>=0.7 and score<0.8:
        zeropseveneight.append(score)
    elif score>=0.8 and score<0.9:
        zeropeightnine.append(score)
    elif score>=0.9 and score<=1.0:
        zeropnineone.append(score)
```

```python
print(len(zerozeropone))
print(len(zeroponetwo))
print(len(zeroptwothree))
print(len(zeropthreefour))
print(len(zeropfourfive))
print(len(zeropfivesix))
print(len(zeropsixseven))
print(len(zeropseveneight))
print(len(zeropeightnine))
print(len(zeropnineone))
```
