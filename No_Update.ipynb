```python
from matplotlib import pyplot as plt
```

```python
iou=[]
with open("C:/Users/vatan/Downloads/groundtruth_query_1.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            iou_str=splitted_line[3]
            #if float(iou_str)>0.3:
            iou.append(float(iou_str))
```

```python
print(iou[0:10])
```

```python
x=list(range(len(iou)))
y=iou
plt.plot(x,y)
plt.xlabel("Frame Number")
plt.ylabel("IoU Score")
plt.title("Query Image Not Updated")
plt.show()
```
