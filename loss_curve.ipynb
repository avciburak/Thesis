```python
all_lines=[]
with open("loss_epoch_2way.txt","r") as f:
    for line in f:
        all_lines.append(line)
```

```python
losses=[]
for line in all_lines:
    losses.append(float(line.split(",")[1]))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(range(len(all_lines)),losses)
plt.show()
```
