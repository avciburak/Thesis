```python
from matplotlib import pyplot as plt
plt.style.use("seaborn")
```

```python
acc_nu=[]
pre_nu=[]
rec_nu=[]
acc=0.0
pre=0.0
rec=0.0
tp=0
tn=0
fp=0
fn=0
with open("C:/Users/vatan/Downloads/no_update.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            tp_str=splitted_line[1]
            tn_str=splitted_line[2]
            fp_str=splitted_line[3]
            fn_str=splitted_line[4]
            tp+=int(tp_str)
            tn+=int(tn_str)
            fp+=int(fp_str)
            fn+=int(fn_str)
            acc=(float(tp)+float(tn))/(float(tp)+float(tn)+float(fn)+float(fp))
            acc_nu.append(acc)
            pre=(float(tp))/(float(tp)+float(fp))
            pre_nu.append(pre)
            rec=(float(tp))/(float(tp)+float(fn))
            rec_nu.append(rec)
```

```python
acc_wu=[]
pre_wu=[]
rec_wu=[]
acc=0.0
pre=0.0
rec=0.0
tp=0
tn=0
fp=0
fn=0
with open("C:/Users/vatan/Downloads/with_update.txt") as f:
    for line in f:
        if line[0]=="f":
            splitted_line=line.split(" | ")
            tp_str=splitted_line[1]
            tn_str=splitted_line[2]
            fp_str=splitted_line[3]
            fn_str=splitted_line[4]
            tp+=int(tp_str)
            tn+=int(tn_str)
            fp+=int(fp_str)
            fn+=int(fn_str)
            acc=(float(tp)+float(tn))/(float(tp)+float(tn)+float(fn)+float(fp))
            acc_wu.append(acc)
            pre=(float(tp))/(float(tp)+float(fp))
            pre_wu.append(pre)
            rec=(float(tp))/(float(tp)+float(fn))
            rec_wu.append(rec)
```

```python
print(len(acc_nu))
print(len(acc_wu))
```

```python
x=list(range(len(acc_wu)))
plt.plot(x,acc_nu[0:3627],color='r',label='no update')
plt.plot(x,acc_wu,color='g',label='with update')
#plt.plot(x,acc_ueff[0:15750],color='g',label='updated in each five frames')
plt.xlabel("Frame Number")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
x=list(range(len(acc_wu)))
plt.plot(x,pre_nu[0:3627],color='r',label='no update')
plt.plot(x,pre_wu,color='g',label='with update')
#plt.plot(x,acc_ueff[0:15750],color='g',label='updated in each five frames')
plt.xlabel("Frame Number")
plt.ylabel("Precision")
plt.legend()
plt.show()
```

```python
x=list(range(len(acc_wu)))
plt.plot(x,rec_nu[0:3627],color='r',label='no update')
plt.plot(x,rec_wu,color='g',label='with update')
#plt.plot(x,acc_ueff[0:15750],color='g',label='updated in each five frames')
plt.xlabel("Frame Number")
plt.ylabel("Recall")
plt.legend()
plt.show()
```
