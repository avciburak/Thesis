```python
classes=[]
with open("class_based_acc_5way_train.txt","r") as f:
    for line in f:
        classes.append(line.split("/")[0])
unique_classes=set(classes)
```

```python
len(unique_classes)
```

```python
all_lines=[]
with open("class_based_acc_5way_train.txt","r") as f:
    for line in f:
        all_lines.append(line)
```

```python
class_based_lines={}
for cl in unique_classes:
    class_lines=[]
    for line in all_lines:
        if line.split("/")[0]==cl:
            class_lines.append(line)
    class_based_lines[cl]=class_lines
```

```python
class_scores={}
for cl in unique_classes:
    total=0
    tp=0
    for line in class_based_lines[cl]:
        splitted=line.split("/")
        tp+=int(splitted[1])
        total+=int(float(splitted[2]))
    class_scores[cl]=[tp,total,float(tp)/float(total)]
        
```
